```python
import numpy as np
import os
import pandas as pd
from pandas import ExcelWriter
import matplotlib.pyplot as plt
import openpyxl
```

Importing data from excel files. Density, isotopes, ECM and chemistry.

```python
dfB16_dens = pd.read_excel('DepthDensity_Bcores_lowRes.xlsx', sheet_name='B16', index=False)
dfB16_d18O = pd.read_excel('Depth_d18O__Bcores.xlsx', sheet_name='B16', index=False)
dfB16_ECM = pd.read_excel('DepthECM__B16_B18_B21.xlsx', sheet_name='B16', index=False)
dfB16_chem = pd.read_excel('DepthChem__B16_B18_B21_B29.xlsx', sheet_name='B16', index=False)
```

Finding w.e. depths of volcanic eruptions (Weissbach 2016).

```python
volcWE_B16 = np.asarray(pd.read_excel('VolcanicEruptions__WE_Depth.xlsx', 'Sheet1', usecols=['B16']))
volcWE_B16use = []

for i in range(len(volcWE_B16)):
    if type(volcWE_B16[i][0]) == float:
        volcWE_B16use.append(volcWE_B16[i][0])
volcWE_B16use = np.asarray(volcWE_B16use)[~np.isnan(volcWE_B16use)]
```

Plotting profiles of interest along with volcanic horizons.

```python
fig_B16, ax_B16 = plt.subplots(2, figsize=(18,10))

ax_B16[0].set(xlabel='Depth', ylabel='Conductivity')
ax_B16[0].plot(dfB16_ECM['depth'],dfB16_ECM['ECM'])

ax_B16[1].set(xlabel='Depth', ylabel='d18O')
ax_B16[1].plot(dfB16_d18O['depth'],dfB16_d18O['d18O'])

for depth in volc_depthIce:
    ax_B16[0].axvline(x=depth,color='k')
    ax_B16[1].axvline(x=depth,color='k')
```
