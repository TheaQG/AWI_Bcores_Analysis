```python
import numpy as np
import os
import pandas as pd
from pandas import ExcelWriter
import matplotlib.pyplot as plt
import openpyxl
#%matplotlib notebook

import matplotlib as mpl
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.size'] = 18
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
coreNames_all = ['B16', 'B17','B18', 'B19', 'B20', 'B21', 'B22', 'B23', 'B26', 'B27', 'B28', 'B29', 'B30'] 
bdot0s = np.array([0.143, 0.115, 0.105, 0.091, 0.096, 0.105, 0.141, 0.116, 0.172, 0.165, 0.174, 0.137, 0.158])
Temp0s = np.array([-32.5, -32.3, -32.3, -30.9, -30.4, -30.1,-29.8, -29.3, -30.3, -30.6, -30.6, -31.6, -31.8])+ 273.15

coreNames_chosen = ['B16', 'B18', 'B26']
IDXcoreNames_chosen = [coreNames_all.index(coreNames_chosen[i]) for i in range(len(coreNames_chosen))]

bdot0s_chosen = bdot0s[IDXcoreNames_chosen]
Temp0s_chosen = Temp0s[IDXcoreNames_chosen]
```

```python
df_dens = []
for i in coreNames_chosen:
    df_dens.append(pd.read_excel('DepthDensity_Bcores_lowRes.xlsx', sheet_name=i, index=False))
```

```python
fig, ax = plt.subplots(figsize = (7,9))
ax.set(xlabel='Density', ylabel='Ice Depth')
ax.invert_yaxis()

for i in range(len(df_dens)):
    ax.plot(df_dens[i]['density'],df_dens[i]['weDepth'], label=coreNames_chosen[i] + ' meas.')
ax.legend()
```

```python
from HL_AnalyticThea_Simple import HL_analyticThea

vals = []

fig2, ax2 = plt.subplots(figsize = (7,9))
ax2.set(xlabel='Density', ylabel='Ice Depth')
ax2.invert_yaxis()

fig3, ax3 = plt.subplots(figsize = (7,9))
ax3.set(xlabel='Density', ylabel='Ice Depth')
ax3.invert_yaxis()

for i in range(len(coreNames_chosen)):
    bdot0 = bdot0s[i]
    Temp0 = Temp0s[i]
    z_vec = np.asarray(df_dens[i]['iceDepth'])
    rhoSurf = df_dens[i]['density'][0]/1000
    
    vals.append(HL_analyticThea(z_vec, rhoSurf, bdot0, Temp0))
    ax2.plot(vals[i][0]*1000, z_vec, label=coreNames_chosen[i] + ' HL')
    
    ax3.plot(df_dens[i]['density'],df_dens[i]['iceDepth'], label=coreNames_chosen[i] + ' meas.')
    ax3.plot(vals[i][0]*1000, z_vec, label=coreNames_chosen[i] + ' HL')

ax2.legend()
ax3.legend()

    
```
