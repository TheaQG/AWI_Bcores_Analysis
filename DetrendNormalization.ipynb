```python
x = np.asarray(d18O_LT) - np.mean(d18O_LT)
d = np.asarray(depth_LT)
x_T = np.concatenate([np.zeros(int(pALT/2)), x, np.zeros(int(pALT/2))])
```

```python
xStart = int(pALT/2)
xEnd = int(len(x_T) - pALT/2)

mus = np.zeros(len(x))
sigmas = np.zeros(len(x))

for i in range(xStart+1,xEnd):
    li = pALT
    sumStart = int(i - li/2)
    sumEnd = int(i+li/2)
    #print(f'Start:{sumStart}')
    #print(li)
    #print(f'Slut:{sumEnd}\n')
    mus[i-xStart-1] = sum(x_T[sumStart:sumEnd]/(li+1))
    sigmas[i-xStart-1] = sum((x_T[sumStart:sumEnd] - mus[i-xStart-1])**2/(li+1))
```

```python
from scipy import signal
d_peaks = signal.find_peaks(x, distance = int(pALT/2), prominence=0.5)[0]


fig,ax = plt.subplots(3,1,figsize=(12,7), sharex=True)
ax[0].plot(d,x,color='k',lw=1)
ax[0].plot(d[d_peaks],x[d_peaks], 'o')
ax[0].plot(d,mus,color='blue',ls='--', lw=1.2)
ax[0].set(ylabel='$\overline{x}$', title=f'Number of peaks: {len(d_peaks)}')#(70.90,80.85))#

#ax[1].plot(d_secs,x_secs)
ax[1].plot(d,x-mus,color='k',lw=1)
ax[1].axhline(y=0,color='blue', ls='--', lw=1.2)
ax[1].plot(d, np.sqrt(2)*np.sqrt(sigmas),color='blue',ls='--', lw=1.2)
ax[1].plot(d, -np.sqrt(2)*np.sqrt(sigmas),color='blue',ls='--', lw=1.2)
ax[1].set(ylabel='$\overline{x} - \overline{\mu}$')


s = (x[:-2]-mus[:-2])/(np.sqrt(2)*np.sqrt(sigmas[:-2]))



S_peaks = signal.find_peaks(s, distance = int(pALT/2), prominence=0.25)[0]

ax[2].plot(d[:-2][S_peaks],s[S_peaks], 'o')
ax[2].plot(d[:-2],s, color='k', lw=1)
ax[2].axhline(y=0,color='blue', ls='--', lw=1.2)
ax[2].axhline(y=-1,color='blue', lw=1.2)
ax[2].axhline(y=1,color='blue', lw=1.2)
ax[2].set(ylabel='$\overline{s}$', xlabel='Depth [m]',title=f'Number of peaks: {len(S_peaks)}')
fig.tight_layout()
#ax[2].plot(d_secs,x_secs,lw=1)
print(len(S_peaks))
```
